```python
from google.colab import files
uploaded = files.upload()
import pandas as pd
import numpy as np
#load Data set
df = pd.read_csv(list(uploaded.keys())[0])
# Show first rows
df.head() # showing first five rows known as headers.
```

```python
# Step 2: Inspect data
print("Data Types:\n", df.dtypes)#Analysing DataTypes
print("\nMissing Values:\n", df.isnull().sum())#Checking for Missing Values
print("\nSummary Statistics:\n", df.describe())#Printing Statics of the Data set.
```

```python
# Step 3: Handle missing values
df = df.fillna(df.mean(numeric_only=True))   # numeric → mean
for col in df.select_dtypes(include='object').columns:  # categorical → mode
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
# Step 4: Detect outliers before handling
for col in df.select_dtypes(include=np.number).columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    print(f"{col}: {len(outliers)} outliers detected")

# Step 4b: Handle outliers (capping)
for col in df.select_dtypes(include=np.number).columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df[col] = np.where(df[col] < lower, lower,
                       np.where(df[col] > upper, upper, df[col]))
```

```python
# Step 5: Save cleaned dataset
df.to_csv("boston_cleaned.csv", index=False)

files.download("boston_cleaned.csv")
```
